### Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model, metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import preprocessing, pipeline
from sklearn.utils.multiclass import unique_labels
```

```python
logreg = linear_model.LogisticRegression(class_weight='balanced')
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_val)


print('Validation', metrics.accuracy_score(y_val, y_pred))
print(classification_report(y_test, logreg.predict(X_test)))
print("Logit confusion matrix: \n\n", confusion_matrix(y_test, logreg.predict(X_test)))
```

```python
logreg = pipeline.Pipeline([
    ("scale", preprocessing.StandardScaler()),
    ("logistic", linear_model.LogisticRegression(C=0.0001, class_weight='balanced'),
    )
])
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_val)

print('Test', metrics.accuracy_score(y_val, logreg.predict(X_val)))
print(classification_report(y_test, logreg.predict(X_test)))
print("Logit confusion matrix: \n\n", confusion_matrix(y_test, logreg.predict(X_test)))
```
